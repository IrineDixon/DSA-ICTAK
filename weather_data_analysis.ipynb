```python
temperature = [30, 32, 31, 29, 28]
humidity = [78,70,68,80,76]
condition = 'sunny'


def avg_temperature(temp_list):
  return sum (temp_list)/len(temp_list)

def avg_humidity(hum_list):
  total = 0
  for h in hum_list:
    total = total + h
  return total / len (hum_list)

print("Average temperature : ", avg_temperature (temperature))
print("Average humidity : ", avg_humidity (humidity))

if condition == 'sunny':
  print('The condition is sunny')
elif condition == 'cloudy':
  print ('The condition is cloudy')
else:
  print ('Unknown Weather')

```
