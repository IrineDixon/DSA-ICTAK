```python
def deposit (amount,balance):
  balance = balance + amount
  print (f"{amount} deposited successfully.")
  print (f"Balance is {balance}")
  return balance


def withdrawal (amount, balance):
  if balance < 50 :
    print ("No suffiecient balance")
  elif amount > balance:
     print ("No suffiecient balance")
  else:
    balance = balance - amount
    print (f"Amount withdrawed : {amount}")
    print (f"Balance is {balance}")
  return balance

def check_balance():
  print(f"current balance is {balance}")

balance = 0.0
balance = deposit (2500, balance)
balance = withdrawal (2000, balance)
balance = withdrawal (600, balance)
check_balance()
```
