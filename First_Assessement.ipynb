2.1 Calculate the total revenue for the store (sum of all quantities ×
price per item).


2.2 Identify all orders where the total order value (quantity × price
per item) exceeds ₹2,000. Store the result as a new list.


2.3 Find the item with the highest total quantity sold across all
orders

```python
total_revenue = 0
high_order = []
item_quantities = {}

for sale in sales_data:
  order_total = sale['quantity'] * sale['price_per_item']
  total_revenue = total_revenue + order_total
  if order_total > 2000:
    high_order.append(sale)
  item_name = sale['item']
  if item_name in item_quantities:
    item_quantities[item_name] = item_quantities[item_name] + sale['quantity']
  else:
    item_quantities [item_name] = sale ['quantity']

best_selling_item = max(item_quantities, key=item_quantities.get)
highest_qty = item_quantities[best_selling_item]




print(f'Total revenue of the store is : {total_revenue:.2f}')
print('--'*20)

print(f'Orders having total order value above 2000 :{len(high_order)} ')
print("Orders:\n")
for item in high_order:
  total_cost = item['quantity'] * item ['price_per_item']
  print(f'Order ID : {item['order_id']}: {item['item']}. Total Cost = {total_cost:.2f}')
print('--'*20)

print(f"Top Selling Item: {best_selling_item} ({highest_qty} units sold)")
```

3.1 Write a function apply_discount(orders, discount_rate) that takes
a list of orders and applies a percentage discount (e.g., 10% = 0.10) to
the price_per_item for each order.

Return a new list of orders (do not modify the original list).

3.2 Explain, with a code example in your script, the difference
between local and global variable scope in Python.


```python
def apply_discount(orders, discount_rate):
    new_discounted_list = []

    for sale in orders:
        new_order = {
            "order_id": sale["order_id"],
            "customer": sale["customer"],
            "item": sale["item"],
            "quantity": sale["quantity"],
            "price_per_item": sale["price_per_item"] * (1 - discount_rate),
            "order_date": sale["order_date"]
        }
        new_discounted_list.append(new_order)

    return new_discounted_list

discounted_sales = apply_discount(sales_data, 0.10)

print(f"{'Order ID':<10} | {'Item':<20} | {'Original Price':<15} | {'New Price (10% off)'}")
print("-" * 75)

for i in range(len(sales_data)):
    original = sales_data[i]
    discounted = discounted_sales[i]

    print(f"{original['order_id']:<10} | {original['item']:<20} | {original['price_per_item']:<15.2f} | {discounted['price_per_item']:.2f}")

#original list is the list given in the first question
```

4.1 Write a custom iterator class OrderIterator to iterate through the
orders, yielding one order at a time.

4.2 Create a dictionary grouping all orders by customer using a
dictionary comprehension.

```python
class OrderIterator:
    def __init__(self, orders):
        self.orders = orders
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.orders):
            order = self.orders[self.index]
            self.index += 1
            return order
        else:
            raise StopIteration

print("Iterating through orders using OrderIterator:")
order_it = OrderIterator(sales_data)

for order in order_it:
    print(f"ID: {order['order_id']} | Item: {order['item']}")
```

```python
unique_customers = {sale['customer'] for sale in sales_data}

customer_groups = {
    cust: [order for order in sales_data if order['customer'] == cust]
    for cust in unique_customers
}

print("\nOrders Grouped by Customer:")
for customer, orders in customer_groups.items():
    items = [o['item'] for o in orders]
    print(f"{customer}: {items}")
```
